```python
# Reading .csv data files

import pandas as pd
import numpy as np
import glob
```

```python
path = r'D:\Ethereum project\ALL DATA' # use your path
all_files = glob.glob(path + "/*.csv")

li = []
date_range=pd.date_range('2016-01-01', '2021-06-01', freq='D')
df_ref=pd.DataFrame(np.zeros(date_range.shape[0]), columns = ['REF'])
df_ref['Date']=date_range
df_ref.set_index('Date', inplace=True)
li.append(df_ref)

for filename in all_files:
#filename=all_files[0]
    if 'Binance_' in filename:
        Coin_Name=filename.split('\\')[-1].split('_')[1][0:-4]
        df = pd.read_csv(filename, index_col='date', header=1, parse_dates=True)
        dft=pd.DataFrame()
        dft['Date']=df.index[-1::-1]
        dft[Coin_Name + '_Price(USDT)']=np.mean([df.open,df.close], axis=0)[-1::-1]
        dft[Coin_Name + '_Volume(USDT)']= df['Volume '+ Coin_Name].values[-1::-1]
        dft[Coin_Name + '_Tradecount']=df['tradecount'].values[-1::-1]
        dft.set_index('Date', inplace=True) 
    if 'export-' in filename:
        Prop_Name=filename.split('\\')[-1].split('export-')[1].split('.')[0]
        df = pd.read_csv(filename, index_col='Date(UTC)', header=0, parse_dates=True)
        dft=pd.DataFrame()
        dft['Date']=df.index
        if 'Value' in df.columns:
            dft[Prop_Name]=df['Value'].values
        else:
            a=set(df.columns).intersection(['Date(UTC)', 'UnixTimeStamp', 'DateTime'])
            b=list(df.columns)
            for i in a:
                b.remove(i)
            dft[Prop_Name]=df[b[0]].values
        dft.set_index('Date', inplace=True)
    if 'BTC Dominance'in filename:
        Prop_Name='BTC-Dominance'
        df = pd.read_csv(filename, index_col='Date', header=0, parse_dates=True)
        dft=pd.DataFrame()
        dft['Date']=df.index
        dft[Prop_Name]=df['Value'].values
        dft.set_index('Date', inplace=True)
    if 'FED-Euro'in filename:
        Prop_Name='USD_EURO'
        df = pd.read_csv(filename, index_col='Date', header=0, parse_dates=True)
        dft=pd.DataFrame()
        dft['Date']=df.index
        dft[Prop_Name]=1/df['Value'].values
        dft.set_index('Date', inplace=True)
    if 'FED-Yuan'in filename:
        Prop_Name='USD_YUAN'
        df = pd.read_csv(filename, index_col='Date', header=0, parse_dates=True)
        dft=pd.DataFrame()
        dft['Date']=df.index
        dft[Prop_Name]=df['Value'].values 
        dft.set_index('Date', inplace=True)
    if 'FRED-WGS7YR'in filename:
        Prop_Name='WGS7YR'
        df = pd.read_csv(filename, index_col='Date', header=0, parse_dates=True)
        dft=pd.DataFrame()
        dft['Date']=df.index
        dft[Prop_Name]=df['Value'].values
        dft.set_index('Date', inplace=True)
    if 'GOLD'in filename:
        Prop_Name='Gold_Price'
        df = pd.read_csv(filename, index_col='Date', header=0, parse_dates=True)
        dft=pd.DataFrame()
        dft['Date']=df.index
        dft[Prop_Name]=df['USD (AM)'].values  
        dft.set_index('Date', inplace=True)
    if 'NASDAQOMX-RCMP'in filename:
        Prop_Name='NASDAQ_Ind'
        df = pd.read_csv(filename, index_col='Trade Date', header=0, parse_dates=True)
        dft=pd.DataFrame()
        dft['Date']=df.index
        dft[Prop_Name]=df['Index Value'].values
        dft.set_index('Date', inplace=True)
    if 'OPEC'in filename:
        Prop_Name='OIL_PRICE'
        df = pd.read_csv(filename, index_col='Date', header=0, parse_dates=True)
        dft=pd.DataFrame()
        dft['Date']=df.index
        dft[Prop_Name]=df['Value'].values
        dft.set_index('Date', inplace=True)
    if 'search_trends'in filename:
        Prop_Name='GOOGLE_TREND'
        df = pd.read_csv(filename, index_col='Date', header=0, parse_dates=True)
        dft=pd.DataFrame()
        dft['Date']=df.index
        dft[Prop_Name]=df['modified daily'].values  
        dft.set_index('Date', inplace=True)
    if 'tweet_volume'in filename:
        Prop_Name='TWEETS'
        df = pd.read_csv(filename, index_col='Date', header=0, parse_dates=True)
        dft=pd.DataFrame()
        dft['Date']=df.index
        dft[Prop_Name]=df['Tweets'].values
        dft.set_index('Date', inplace=True)
    li.append(dft)

from functools import reduce
dfm = reduce(lambda df1,df2: pd.merge(df1,df2,how='outer', left_index=True, right_index=True), li)
dftotal=dfm.loc['2016-1-1':'2021-6-1']
dftotal.to_csv('Merge.csv')
```
